## Conversió de dades

Ara que ja no tenim NaNs, anem a arreglar les columnes que tenguin un contingut que es pugui considerar categòric i que poguem mapejar a un valor numèric.

```python
df
```

Ara duim a terme les declaracions de les funcions de neteja de dades per aquestes dades de tipus diferent a int o float que són fàcilment convertibles a aquests tipus llevant les unitats de mesura ("Value", "Release Clause", "Wage", "Height", "Weight") i calculant les expressions calculables ("LS" fins a "RB"):

```python
#Per "Value", "Release Clause" i "Wage"
def money_to_float(x): 
    if type(x) == float or type(x) == int:
        return float(x)
    else:
        ret_val = 0.0
        x = x.replace("€", "")
        if 'K' in x:
            if len(x) > 1:
                ret_val = float(x.replace('K', ''))
            ret_val = ret_val *1000
        if 'M' in x:
            if len(x) > 1:
                ret_val = float(x.replace('M', ''))
            ret_val = ret_val * 1000000.0
        return ret_val

#Per l'altura
def height_to_float(x): 
    return float(x.replace("'", "."))

#Pel pes
def weight_to_float(x): 
    return float(x.replace("lbs", ""))

#Pels valors de forma [int]+[int]
def suma_to_float(x): 
    ll = x.split("+")
    return float(ll[0]) + float(ll[1])
```

Apliquem les funcions als camps esperats:

```python
df["Value"] = df["Value"].apply(money_to_float)
df["Wage"] = df["Wage"].apply(money_to_float)
df["Release Clause"] = df["Release Clause"].apply(money_to_float)

df["Height"] = df["Height"].apply(height_to_float)

df["Weight"] = df["Weight"].apply(weight_to_float)

sub = ["LS", "ST", "RS", "LW", "LF", "CF", "RF", "RW", "LAM", "CAM", "RAM", "LM", "LCM", "CM", "RCM", "RM", "LWB", "LDM", 
       "CDM", "RDM", "RWB", "LB", "LCB", "CB", "RCB", "RB"]
for col in sub:
    df[col] = df[col].apply(suma_to_float)

df
```

Si ens fixam bé "Work Rate" indica la capacitat ofensiva i defensiva del jugador amb format "capacitat ofensiva"/"capacitat defensiva". Es tracten d'unes dades categòriques. També es pot apreciar que es poden separar en dues columnes aquestes dades, una per la capacitat ofensiva i una altra per la defensiva. També podem associar un valor major a una major puntuació d'ofensivitat o defensivitat:
* Low es mapeja a 1
* Medium es mapeja a 2
* High es mapeja a 3

Apliquem les operacions necessàries

```python
def to_float_wr(x):
    x = x.replace(" ","")
    if x == "Low":
        return 1.0
    elif x == "Medium":
        return 2.0
    elif x == "High": 
        return 3.0
    return 0.0

def offensives_defensives(wr, ofensives, defensives):
    for x in wr:
        ll = x.split("/")
        ofensives.append(to_float_wr(ll[0]))
        defensives.append(to_float_wr(ll[1]))
        
ofensives = []
defensives = []
offensives_defensives(df["Work Rate"], ofensives, defensives)
df.insert(13, "Defensivitat", defensives, False)
df.insert(13, "Ofensivitat", ofensives, False)
df = df.drop(['Work Rate'], axis=1)

df
```

Ara anem a aplicar <a href = "https://es.wikipedia.org/wiki/One-hot">*One Hot Encoding*</a> sobre les dades categòriques que queden per processar -Club i Position- (O ets del FC Barcelona o no, si es discretitza el club assignant un nombre arbitrari a cada club tenim clubs que tenen un valor gros assignat i altres amb un de petit. Si prenem una transformació quantitativa amb aquesta variable respecte la regressió lieal l'únic que aconseguiriem és generar renou. És millor fer una classificació binària: pertany/no pertany. Això també s'aplica a la resta de dades categòriques esmentades anteriorment):

```python
clb = df.pop("Club")
df = pandas.concat([df.reset_index(drop=True), pandas.get_dummies(clb, prefix='clb').reset_index(drop=True)], axis=1, sort=False)

pos = df.pop("Position")
df = pandas.concat([df.reset_index(drop=True), pandas.get_dummies(pos, prefix='pos').reset_index(drop=True)], axis=1, sort=False)

df
```

# Entranament del model

És hora de dur a terme l'entrenament del model.

En primer lloc separam la variable "Value" (la que volem preveure) de la resta del conjunt de dades per entrenar el model.

```python
val = df.pop("Value")
```

Una vegada fet això anem a fer un split separant les dades a una raó del 75% per l'entrenament del model i un 25% per les proves damunt aquest.

```python
X_train, X_test, y_train, y_test = train_test_split(df, val, test_size=0.25, random_state=42)
```

Una vegada ho tenim tot preparat és hora d'entrenar el nostre model.

```python
reg = linear_model.LinearRegression().fit(X_train, y_train)
```

Anem a veure la qualitat de les nostres prediccions. Obtendrem una mètrica <a href = "https://en.wikipedia.org/wiki/Coefficient_of_determination#:~:text=In%20case%20of%20a%20single,predictor%20and%20the%20response%20variable."> $R^2$ </a>, el coeficient de la qual determina la qualitat del nostre model per preveure els resultats reals. El millor valor possible és 1, i el menor és 0.

Anem a realitzar prediccions amb les dades d'entrada que haviem separat per llavors comparar-les amb les sortides originals.

```python
preds = reg.predict(X_test)
r2_score(preds, y_test)
```

Com podem veure, el resultat és favorable perquè s'aproxima a 1. Es determina que existeix una correlació sustanciada entre les prediccions del nostre model i les dades reals.
